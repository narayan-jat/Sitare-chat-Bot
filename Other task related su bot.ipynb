```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('./SUCHAT_EMB - Sheet1.csv')
questions = df["question"]
answers = df["answer"]
```

```python
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
question_embeddings = model.encode(questions)
answer_embeddings = model.encode(answers)
```

```python
import numpy as np
def answer_question(question):
    asked_question_embeddings = model.encode(question)
    # similarity_scores = cosine_similarity([asked_question_embeddings], question_embeddings)
    similarity_scores = cosine_similarity([asked_question_embeddings], answer_embeddings)
    most_similar_answer_index = similarity_scores.argmax()
    # most_similar_question_index = similarity_scores.argmax()
    # most_similar_answer_index = cosine_similarity([question_embeddings[most_similar_question_index]], answer_embeddings).argmax()
    return answers[most_similar_answer_index]
```

```python
question = "what is sitare"
answer = answer_question(question)
print("Answer:", answer)
```

```python
from transformers import AutoTokenizer, AutoModel

# There are a lot of BERT based models available on HuggingFace,
# and you have to pick one that is suitable for you.
BERT_Model = "bert-base-uncased"

# Initialize the tokenizer, which will break the input text into tokens (sub-words)
tokenizer = AutoTokenizer.from_pretrained(BERT_Model)

# Initialize the LLM that will take the tokens and output the embeddings
model = AutoModel.from_pretrained(BERT_Model)
```

```python

# Tokenize the input text
def tokenize_questions():
    questions_tokens = []
    for i in range(len(questions)):
        tokens = tokenizer.encode_plus(questions[i], max_length=128, truncation=True,
                                padding='max_length', return_tensors='pt')
        questions_tokens.append(tokens)
        break
    return questions_tokens

def hidden_state_embeddings(questions_tokens):
    embeddings = []
    for i in range(len(questions_tokens)):
        outputs = model(**questions_tokens[i])
        embeddings.append(outputs.last_hidden_state)
        break
    return embeddings

def pooler_output_embeddings(questions_tokens):
    embeddings = []
    for i in range(len(questions_tokens)):
        outputs = model(**questions_tokens[i])
        embeddings.append(outputs.pooler_output)
    return embeddings


questions_tokens = tokenize_questions()
hidden_state_embedding = hidden_state_embeddings(questions_tokens)
# pooler_output_embedding = pooler_output_embeddings(questions_tokens)

print(len(hidden_state_embedding))
# print(len(pooler_output_embedding))
```
